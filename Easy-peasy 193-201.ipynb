```python
def fill_value(height, width, value):
  return [[value for row_value in range(width)] for row in range(height)]

fill_value(4, 2, 'a')
```

```python
def trace(matrix):
  return sum([row[idx] for idx, row in enumerate(matrix)])

trace([[1]])
```

```python
def transpose(array):
  return [[array[row_number][idx] for row_number in range(len(array))] for idx in range(len(array[0]))]

transpose([[1, 2, 3, 4], [5, 6, 7, 8]])
```

```python
def max_prob(array):
  return [max(row) for row in array]

max_prob([[0.3, 0.4, 0.2, 0.1], [0.0, 0.1, 0.7, 0.2]])
```

```python
def detect_class(array):
  return [[1 if max(row)==row_index else 0 for row_index in row] for row in array]

detect_class([[0.3, 0.4, 0.2, 0.1], [0.0, 0.1, 0.7, 0.2], [0.0, 0.3, 0.3, 0.4]])
```

```python
def dot_product(list1, list2):
  return sum([list1[idx]*list2[idx] for idx in range(len(list1))])

dot_product([1, 2], [5, 2])
```

```python
def count_none(list_):
  return len(list(filter(lambda x: x==None, list_)))

count_none([1, None, None, 5, None, 2])
```

```python
def top_n(items, n):
  return sorted(items, reverse=True)[:n]

top_n([4, 5, 2, 9, 5, 2, 8, 2, 8, 10], 3)
```
